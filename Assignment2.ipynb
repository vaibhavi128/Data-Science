```python

import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Display the first few rows of the dataset
print("First five rows of the Iris dataset:")
print(iris_df.head())

# Basic descriptive statistics using the describe() function
print("\nBasic Descriptive Statistics:")
print(iris_df.describe())

# Calculating Quartiles
print("\nQuartiles (25%, 50%, 75%):")
print(iris_df.quantile([0.25, 0.5, 0.75]))

# Mean of each column
print("\nMean of each column:")
print(iris_df.mean())

# Median of each column
print("\nMedian of each column:")
print(iris_df.median())

# Mode of each column
print("\nMode of each column:")
print(iris_df.mode().iloc[0])  # Using .iloc[0] to get the first row, as mode() returns a DataFrame

# Variance of each column
print("\nVariance of each column:")
print(iris_df.var())

# Standard Deviation of each column
print("\nStandard Deviation of each column:")
print(iris_df.std())

# Range (max - min) for each column
print("\nRange of each column (Max - Min):")
print(iris_df.max() - iris_df.min())
```
