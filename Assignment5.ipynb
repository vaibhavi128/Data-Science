```python
df.isnull().sum()
# Message column have Null value.
```

```python
# Convert 'Message' column to object type
df['Message'].astype('object')

# Now fill missing values with "no message"
df['Message'].fillna("no message")
```

```python
df.isnull().sum()
# there is no any column that have null values
```

```python
#  we will give date time format for "Start_Date" Column
pd.to_datetime(df['Start_Date'], errors='coerce')
```

```python
# remove duplicate value
df.drop_duplicates()
```

```python
# one-hot encoding(categorical into numerical format )
pd.get_dummies(df, columns=['Geo Type Name', 'Geo Place Name','Name'], drop_first=True)
```

```python
# Extract the year from 'Start_Date'
df['Start_Date'].dt.year
```

```python
# Aggregation and Grouping
# Calculate monthly or yearly averages for each pollutant and location to observe trends over time.

df.groupby(['Year', 'Geo Place Name', 'Name']).agg({'Data Value': 'mean'}).reset_index()
```
