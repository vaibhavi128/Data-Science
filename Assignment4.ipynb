**Implement normal distribution in python and visualize it for Mean =100, standard _deviation =4, dataset size=100000**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 100
std_dev = 4
dataset_size = 100000

# Step 1: Generate the data using normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=dataset_size)

# Step 2: Visualize the normal distribution using seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=50, kde=True, color='gray')

# Adding labels and title
plt.title(f'Normal Distribution (Mean={mean}, Std Dev={std_dev}, Size={dataset_size})', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()


```

```python
# Step 1: Generate the data using the standard normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=dataset_size)

# Step 2: Visualize the standard normal distribution using seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=50, kde=True, color='green')

# Adding labels and title
plt.title(f'Standard Normal Distribution (Mean={mean}, Std Dev={std_dev}, Size={dataset_size})', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

```python

import scipy.stats as stats
# Step 2: Create a Q-Q plot

plt.figure(figsize=(8, 6))
stats.probplot(data, dist="norm", plot=plt)

# Adding title and labels
plt.title(f'Q-Q Plot (Mean={mean}, Std Dev={std_dev}, Size={dataset_size})', fontsize=14)
plt.xlabel('Theoretical Quantiles', fontsize=12)
plt.ylabel('Sample Quantiles', fontsize=12)

# Show the plot
plt.show()
```

```python
# Step 2: Shapiro-Wilk Test
shapiro_stat, shapiro_p = stats.shapiro(data)
print(f'Shapiro-Wilk Test: Statistic={shapiro_stat}, p-value={shapiro_p}')

# Step 3: Kolmogorov-Smirnov Test
ks_stat, ks_p = stats.kstest(data, 'norm', args=(mean, std_dev))
print(f'Kolmogorov-Smirnov Test: Statistic={ks_stat}, p-value={ks_p}')
```
